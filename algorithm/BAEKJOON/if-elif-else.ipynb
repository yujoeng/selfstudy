**백준 1330번 두 수 비교하기 **

두 정수 A와 B가 주어졌을 때, A와 B를 비교하는 프로그램을 작성하시오. 

배운 점 : if - elif - else문을 활용해서 a 아니면 b 아니면 c 의 개념을 알 수 있었음 

주의할 점 :  반복 for문을 사용하지 않아도 된다는 점과 '=='로 입력하는 방법, 들여쓰기의 위치를 확실하게 할 것!!!! 

내 풀이 
```
a, b = map(int, input().split())

if a > b :
    print('>')
    
elif a < b :
    print('<')
    
else:
    print('==')
```

다른 사람 풀이 
# 삼항 연산자 코드 
a, b = map(int, input().split())

print('>') if a > b else print('<') if a < b else print('==')

#코드 설명 
조건식 1이 참일 때 값 if 조건식 1 else [조건식2가 참일 때 값 if 조건식2 else 조건식이 모두 거짓일 때 값]

## 내 코드

    print('1')

배운 점 : 연산자 입력하는 방법 
# [a%4 == 0]
n의 배수를 입력하기 위해서는 나머지 연산자 (%)를 이용하여 표현한다. 추가로 4의 배수는 연도로 입력받은 수를 4로 나누었을 때 나머지가 0인 값으로 표현. 
# 비교연산자 [== 같다,  != 같지 않다 ]
100의 배수가 아닐 때 또는 400의 배수일 때 라는 조건이 주어져서 같다와 같지않음의 비교연산자를 이용해서 입력했다. 
# 비교연산자 [and 와 or]
"4의 배수이면서, 100의 배수가 아닐 때 또는 400의 배수일 때" 라는 조건이 주어져서 4의 배수 조건과 100의 배수 조건을 and 로 연결하고 한 괄호에 넣어서 입력했다.  또는 400의 배수일 때 를 충족시키기 위해 or 연산자를 활용함.
